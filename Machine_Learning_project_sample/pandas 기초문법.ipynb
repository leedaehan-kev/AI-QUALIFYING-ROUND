```python
df2
```

```python
df2.info()
```

### 운동량 컬럼은 숫자형 컬럼같이 보이지만 object로 돼있다. 
### 종종 이런 현상이 발생하는데, 이런 현상을 만들어낸 이유는 분명 존재합니다!

### 원인 파악을 위해서 일단 pd.to_numeric()으로 변환을 시도합니다.

```python
pd.to_numeric(df2['운동량'])
```

### 숫자형으로 바꿀 때 NaN이나  문자열이 존재할 때 숫자형변환이 실패한다.
### 이때는 errors=옵션값을 바꾸어서 해결하자
####  errors='coerce' : 잘못된 문자열은 NaN으로 치환
####  errors='ignore' : 잘못된 문자열이 숫자로 변환이 안되고 무시하기 때문에 전체컬럼의 dtype이 object로 그대로 남아있다.

```python
pd.to_numeric(df2['운동량'], errors='coerce')
```

```python
pd.to_numeric(df2['운동량'], errors='ignore')
```

```python
#재대입으로 마무리 
df2['운동량'] = pd.to_numeric(df2['운동량'], errors='coerce')
```

# pd.cut() - 간격 동일하게 구간 나누기(binning)

```python
df2.head()
```

```python
df2.describe()
```

#### 운동량은 범위가 매우 넓다. 최소값은 0인데 최대값은 1.639361e+08 이다.
#### 따라서 운동량을 기준으로 데이터를 10개그룹으로 분류하자

```python
df2['운동량_cut'] = pd.cut(df2['운동량'],bins=10)
```

```python
df2['운동량_cut'].value_counts()
```

### 분포를 보니깐 대부분이 첫구간에 쏠려있다. 
## 즉, pd.cut()은 최소에서 최대구간을 지정한 bin만큼 동일하게 분할
## 이상치가 있는 경우에 안좋다

# pd.qcut() - 동일한 갯수를 갖도록 구간 분할 

```python
df2['운동량_qcut'] = pd.qcut(df2['운동량'],q=10)
```

```python
df2['운동량_qcut'].value_counts()
```
