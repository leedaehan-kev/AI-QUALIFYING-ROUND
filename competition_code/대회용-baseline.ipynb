```python
train  = pd.read_csv('competition_data/train.csv')
test  = pd.read_csv('competition_data/test.csv')
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_colwidth', 100)
pd.set_option('display.max_columns', 100)
```

```python
train = train.drop(['index', 'country'],axis = 1)
test =test.drop(['index', 'country'],axis = 1)
```

```python
clf = setup(data = train, target = 'nerdiness')
```

```python
best_3 = compare_models(sort = 'AUC', n_select = 3)
```

```python
et = create_model('et')
rf = create_model('rf')
```

```python
blended = blend_models([et,rf], weights = [0.8,0.2], fold = 10, method = 'soft')
```

```python
tuned_blender = tune_model(blended,optimize='AUC',n_iter=50)
```

```python
pred_holdout = predict_model(tuned_blender)
```

```python
final_model = finalize_model(tuned_blender)
```

```python
predictions = predict_model(final_model, data = test)
```
